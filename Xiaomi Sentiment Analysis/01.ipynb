## 1. Data Mining (Ekstraksi Komentar dari Youtube)

Langkah pertama Untuk melakukan analisis sentimen adalah memperoleh data teks berupa komentar dari video YouTube terkait. YouTube tidak menyediakan fitur ekspor komentar secara langsung, sehingga kita perlu melakukan web scraping menggunakan [YouTube API](https://developers.google.com/youtube/v3) yang resmi dari Google.

```python
def extract_youtube_comments(video_id, api_key, max_comments=2500):
    """
    Mengambil komentar dari YouTube dengan paginasi dan error handling.
    """
    youtube = build('youtube', 'v3', developerKey=api_key)
    comments = []
    next_page_token = None
    fetched = 0
    
    # Progress bar untuk estimasi
    pbar = tqdm(total=max_comments, desc=f"Scraping {video_id}", leave=False)

    try:
        while fetched < max_comments:
            request = youtube.commentThreads().list(
                part='snippet',
                videoId=video_id,
                maxResults=100,
                pageToken=next_page_token,
                textFormat='plainText',
                order="relevance" # Mengambil komentar terpopuler/relevan
            )
            response = request.execute()

            for item in response.get('items', []):
                snippet = item['snippet']['topLevelComment']['snippet']
                comments.append({
                    "video_id": video_id,
                    "author": snippet.get('authorDisplayName'),
                    "text": snippet.get('textDisplay', ''),
                    "like_count": snippet.get('likeCount', 0),
                    "published_at": snippet.get('publishedAt')
                })
                fetched += 1
                pbar.update(1)
                
                if fetched >= max_comments:
                    break

            next_page_token = response.get('nextPageToken')
            if not next_page_token:
                break
                
    except Exception as e:
        print(f"Error pada video {video_id}: {e}")
    finally:
        pbar.close()

    return pd.DataFrame(comments)
```

```python
# Eksekusi Ekstraksi Data Multi-Channel
all_dfs = []
print("Memulai Ekstraksi Data Multi-Channel")

for video in VIDEO_LIST:
    print(f"-> Processing: {video['channel']}")
    df_vid = extract_youtube_comments(video["id"], YOUTUBE_API_KEY, max_comments=3000) # Ambil max 3000 per video
    
    if not df_vid.empty:
        df_vid["source_channel"] = video["channel"]
        df_vid["video_title"] = video["title"]
        all_dfs.append(df_vid)
        # Simpan raw per video (Backup)
        filename = f"{OUTPUT_DIR}/raw_{video['channel'].replace(' ', '_')}.csv"
        df_vid.to_csv(filename, index=False)

# Gabungkan semua data
if all_dfs:
    df_raw = pd.concat(all_dfs, ignore_index=True)
    print(f"\n TOTAL DATA TERKUMPUL: {len(df_raw):,} Komentar")
    df_raw.to_csv(f"{OUTPUT_DIR}/combined_raw_data.csv", index=False)
else:
    print("Gagal mengambil data.")
```

Oleh karena itu, langkah selanjutnya akan difokuskan pada:
- Pengelompokan panjang komentar `len_bucket`, untuk memisahkan komentar yang terlalu pendek (potensial noise) dan komentar panjang (potensial ambiguous sentiment).
- Analisis distribusi per `channel_source`, untuk memastikan tidak terjadi bias gaya bahasa atau audience antar reviewer.

Pendekatan ini memastikan bahwa pipeline sentiment analysis tidak hanya akurat secara statistik, tetapi juga robust, explainable, dan adil terhadap variasi data nyata di lapangan.

### 2.2 Pengelompokan panjang komentar (`len_bucket`)

```python
# 1) Menghitung panjang komentar dalam jumlah kata
df_raw["text_length"] = (
    df_raw["text"]
    .fillna("")                      # jaga-jaga ada NaN
    .astype(str)
    .str.strip()
    .str.split()
    .str.len()
)

# 2) Mendefinisikan bucket (rule-based)
# Threshold disusun berdasarkan statistik awal: Q1=6, median=8, Q3=13, max=103
bins = [-1, 2, 5, 13, 30, np.inf]
labels = ["very_short(0-2)", "short(3-5)", "medium(6-13)", "long(14-30)", "very_long(31+)"]

df_raw["len_bucket"] = pd.cut(df_raw["text_length"], bins=bins, labels=labels)

# 3) Quick check distribusi
bucket_dist = (
    df_raw["len_bucket"]
    .value_counts(dropna=False)
    .rename_axis("len_bucket")
    .reset_index(name="count")
)
bucket_dist["pct"] = (bucket_dist["count"] / len(df_raw) * 100).round(2)

display(bucket_dist)
```

Kesimpulan mengenai pengelompokan panjang komentar:

| Zona| Proporsi | Makna |

| Medium | 53% | Merepresentasikan opini singkat namun informatif |

## 3. Text Preprocessing

Berdasarkan hasil Raw Data EDA, komentar YouTube menunjukkan variasi panjang teks yang ekstrem, penggunaan slang dan emoji yang intens, serta perbedaan gaya bahasa antar channel reviewer. Kondisi ini membuat data mentah berisiko mengandung noise tinggi dan bias linguistik spesifik channel jika langsung digunakan untuk pemodelan. Oleh karena itu, tahap Text Preprocessing diperlukan sebagai jembatan untuk menstandarkan teks, mengurangi noise non-informatif, dan menjaga sinyal emosional yang relevan, sehingga data siap digunakan untuk analisis sentimen yang lebih stabil dan generalizable.

```python
# Slang dictionary untuk mengganti slang ke bentuk formal
slang_dict = {
    # Kata Umum
    "gak": "tidak", "ga": "tidak", "gk": "tidak", "nggak": "tidak", "kagak": "tidak",
    "nda": "tidak", "ndak": "tidak", "tdk": "tidak",
    "gabisa": "tidak bisa", "gak bisa": "tidak bisa",
    "gabole": "tidak boleh",

    "jgn": "jangan", "jan": "jangan",
    "blm": "belum", "belom": "belum",
    "udh": "sudah", "udah": "sudah", "sdh": "sudah",
    "kalo": "kalau", "klo": "kalau",
    "krn": "karena", "karna": "karena",
    "tp": "tapi", "tpi": "tapi",
    "yg": "yang",

    # Kata Ganti
    "gw": "saya", "gue": "saya", "gua": "saya", "ane": "saya",
    "lu": "kamu", "lo": "kamu", "elu": "kamu",
    "gan": "teman", "bro": "teman", "sis": "teman",
    "bgt": "sangat", "banget": "sangat",
    "parah": "sangat",
    "fix": "pasti",

    # Emosi & Reaksi
    "ngakak": "tertawa",
    "wkwk": "tertawa",
    "anjir": "terkejut",
    "buset": "terkejut",
    "baper": "emosional",
    "ilfil": "tidak suka",

    # Review Gadget ‚Äì POSITIF
    "mantap": "bagus",
    "keren": "bagus",
    "kece": "bagus",
    "juara": "terbaik",
    "worth it": "layak beli",
    "smooth": "lancar",
    "ngebut": "cepat",
    "gahar": "kuat",
    "badak": "awet",
    "rekomen": "direkomendasikan",

    # Review Gadget ‚Äì NEGATIF
    "ampas": "buruk",
    "lemot": "lambat",
    "lag": "lag",
    "panas": "panas",
    "boros": "boros",
    "overheat": "terlalu panas",
    "kemahalan": "terlalu mahal",
    "zonk": "mengecewakan",
    "bug": "bug",
    "error": "error",
    "gimmick": "tipuan pemasaran",
    "skip": "jangan beli",

    # Brand & Gadget
    "ipon": "iphone",
    "aipon": "iphone",
    "siomay": "xiaomi",
    "samsul": "samsung",
    "pocong": "poco",

    # Istilah Teknis
    "hp": "ponsel",
    "hape": "ponsel",
    "batre": "baterai",
    "ngecas": "mengisi daya",
    "cas": "isi daya",
    "sot": "waktu layar menyala",
    "hz": "hertz",

    # Spam & Judol
    "judol": "judi online",
    "slot": "judi slot",
    "gacor": "mudah menang",
    "maxwin": "kemenangan besar",
    "depo": "deposit",
    "wd": "penarikan dana",
}
```

```python
# Pre-compile regex biar lebih cepat + rapi
RE_URL_MENTION = re.compile(r"http\S+|www\S+|@\w+")
RE_REPEAT = re.compile(r"(.)\1{2,}")  # >= 3 repetisi
RE_NON_TEXT = re.compile(r"[^a-z0-9\s]")  # Keep angka, agar "5g", "120hz" tidak hilang
RE_MULTI_SPACE = re.compile(r"\s+")

# Menangani phrase slang terlebih dahulu (multi-word), baru token-level
# contoh: "gak bisa", "worth it"
PHRASE_SLANG = {k: v for k, v in slang_dict.items() if " " in k}
WORD_SLANG = {k: v for k, v in slang_dict.items() if " " not in k}

# Build regex phrase dengan boundary agar tidak nyangkut substring aneh
# Mengurutkan phrase terpanjang terlebih dahulu biar gak ketimpa
phrase_patterns = sorted(PHRASE_SLANG.keys(), key=len, reverse=True)
RE_PHRASES = [
    (re.compile(rf"\b{re.escape(p)}\b"), PHRASE_SLANG[p])
    for p in phrase_patterns
]

def clean_text_advanced(text, keep_numeric=True, max_repeat=2):
    if not isinstance(text, str) or not text.strip():
        return ""

    text = text.lower().strip()

    # 1) Mendeteksi dan mengekstrak emoji dari teks agar bisa diproses secara terstruktur.
    text = demoji.replace(text, " ")

    # 2) Hapus URL & Mention
    text = RE_URL_MENTION.sub(" ", text)

    # 3) Normalisasi huruf berulang: "baaaagus" -> "baagus" (maks 2)
    def _reduce(m):
        ch = m.group(1)
        return ch * max_repeat
    text = RE_REPEAT.sub(_reduce, text)

    # 4) Replace phrase slang terlebih dahulu sebelum simbol dibuang
    # biar "gak bisa" / "worth it" terpakai
    for pattern, repl in RE_PHRASES:
        text = pattern.sub(repl, text)

    # 5) Hapus simbol
    if keep_numeric:
        text = RE_NON_TEXT.sub(" ", text)
    else:
        text = re.sub(r"[^a-z\s]", " ", text)

    # 6) Token slang (single word)
    words = text.split()
    fixed_words = [WORD_SLANG.get(w, w) for w in words]
    text = " ".join(fixed_words)

    # 7) Hapus spasi ganda
    return RE_MULTI_SPACE.sub(" ", text).strip()
```

```python
# Eksekusi Cleaning
print("Proses Cleaning Text:")
tqdm.pandas()
df = df_raw.copy()
df["clean_text"] = df["text"].progress_apply(clean_text_advanced)

# Filter: Hapus komentar terlalu pendek (< 4 kata) karena biasanya tidak bermakna
MIN_WORDS = 4
df = df[df["clean_text"].str.split().str.len() >= MIN_WORDS].reset_index(drop=True)
```

### 3.1 Keyword Signal Audit (PRE-SENTIMENT CHECK)

Sebelum menjalankan baseline sentiment classification, dilakukan keyword signal audit untuk memastikan bahwa proses preprocessing tidak menghilangkan sinyal sentimen eksplisit yang akan digunakan pada tahap keyword boosting. Tahap ini berfungsi sebagai sanity check terhadap distribusi keyword positif dan negatif, serta untuk mengidentifikasi komentar ambigu yang berpotensi digunakan pada active learning.

```python
# 1) Mendefinisikan keyword lexicon
POS_KW = [
    # core positive
    "bagus", "baik", "mantap", "mantep", "mantul", "keren", "kece", "cakep",
    "oke", "ok", "sip", "top", "solid", "nice",
    "terbaik", "juara", "the best",

    # recommendation / purchase approval
    "rekomen", "recommended", "direkomendasikan",
    "layak beli", "worth", "worth it", "value", "value for money", "vfm",
    "murah", "harganya pas", "harga oke", "harga bagus", "sepadan",

    # performance praise
    "lancar", "smooth", "responsif", "cepat", "kencang", "ngebut", "ngacir",
    "stabil", "kenceng",

    # durability / quality
    "awet", "tahan lama", "kokoh", "rapi", "build quality bagus",

    # hype / affirmation phrases (umum di YouTube)
    "gas", "gaskeun", "no debat", "auto beli", "fix beli", "wajib beli",
    "mantap jiwa", "gokil", "gila bagus"
]

NEG_KW = [
    # core negative
    "buruk", "jelek", "payah", "parah", "ampas", "zonk",
    "mengecewakan", "kecewa", "ngaco", "gagal",

    # purchase rejection / warning
    "jangan beli", "jangan dibeli", "skip", "mending jangan",
    "tidak worth", "ga worth", "gak worth", "nggak worth",
    "tidak layak", "rugi", "nyesel", "kapok",

    # performance issues
    "lambat", "lemot", "lag", "ngelag", "stutter", "ngadat", "loading lama",
    "crash", "force close", "drop fps",

    # software/quality issues
    "bug", "banyak bug", "error", "glitch", "masalah", "bermasalah",

    # thermal / battery issues
    "panas", "cepat panas", "overheat", "terlalu panas",
    "boros", "baterai boros", "baterai cepat habis", "drop",

    # ads/bloatware complaints
    "iklan", "iklan banyak", "bloatware", "aplikasi bawaan",

    # pricing negative
    "mahal", "kemahalan", "overprice", "tidak sepadan", "ga sepadan", "ketinggian",

    # marketing skepticism
    "gimmick", "tipuan", "pemasaran doang", "cuma gimmick"
]

AMBIG_KW = [
    # neutral/uncertain assessment
    "standar", "biasa", "normal", "lumayan", "cukup", "cukup oke", "cukup bagus",
    "okelah", "pas",

    # indecision / comparison talk
    "mending", "banding", "dibanding", "vs", "pilih", "opsi", "alternatif",
    "nunggu", "tunggu", "nanti", "lihat dulu",

    # question-style / seeking advice
    "gimana", "bagaimana", "worth kah", "rekomen kah", "saran", "minta saran",

    # ‚Äúmixed vibe‚Äù words (sering muncul tapi polar lemah)
    "kurang", "agak", "kayaknya", "mungkin"
]
```

```python
# 2) Helper untuk compile pattern "whole word/phrase"
def compile_phrase_pattern(phrases):
    # urutkan yang panjang dulu biar phrase "jangan beli" kebaca sebelum "beli"
    phrases = sorted(set(phrases), key=len, reverse=True)
    # \b aman untuk kata; untuk phrase (ber-spasi) tetap aman karena ada boundary di ujung2
    pattern = r"|".join([rf"\b{re.escape(p)}\b" for p in phrases])
    return re.compile(pattern, flags=re.IGNORECASE)

pos_pat = compile_phrase_pattern(POS_KW)
neg_pat = compile_phrase_pattern(NEG_KW)
amb_pat = compile_phrase_pattern(AMBIG_KW)
```

```python
# 3) Fungsi ekstrak keyword yang match per komentar
def extract_matches(text, pattern):
    if not isinstance(text, str) or not text.strip():
        return []
    return list(dict.fromkeys(pattern.findall(text)))  # unique preserve order

df_kw = df.copy()

df_kw["pos_hits"] = df_kw["clean_text"].apply(lambda x: extract_matches(x, pos_pat))
df_kw["neg_hits"] = df_kw["clean_text"].apply(lambda x: extract_matches(x, neg_pat))
df_kw["amb_hits"] = df_kw["clean_text"].apply(lambda x: extract_matches(x, amb_pat))

df_kw["pos_hit_count"] = df_kw["pos_hits"].apply(len)
df_kw["neg_hit_count"] = df_kw["neg_hits"].apply(len)
df_kw["amb_hit_count"] = df_kw["amb_hits"].apply(len)

df_kw["has_pos_kw"] = df_kw["pos_hit_count"] > 0
df_kw["has_neg_kw"] = df_kw["neg_hit_count"] > 0
df_kw["has_any_kw"] = df_kw[["has_pos_kw", "has_neg_kw"]].any(axis=1)

df_kw["cooccur_pos_neg"] = df_kw["has_pos_kw"] & df_kw["has_neg_kw"]  # indikasi campuran
```

```python
# 4) Summary distribusi & coverage
total = len(df_kw)
summary = pd.DataFrame({
    "Metric": [
        "Total komentar",
        "Komentar mengandung POS keyword",
        "Komentar mengandung NEG keyword",
        "Komentar mengandung POS atau NEG keyword",
        "Komentar POS & NEG muncul bersamaan",
    ],
    "Count": [
        total,
        int(df_kw["has_pos_kw"].sum()),
        int(df_kw["has_neg_kw"].sum()),
        int(df_kw["has_any_kw"].sum()),
        int(df_kw["cooccur_pos_neg"].sum()),
    ]
})
summary["%"] = (summary["Count"] / total * 100).round(2)

display(summary)
```

Analisis distribusi keyword sentiment menunjukkan bahwa hanya sekitar 16% komentar mengandung ekspresi sentimen eksplisit berbasis kata kunci positif atau negatif. Hal ini mengindikasikan bahwa mayoritas opini pengguna disampaikan secara implisit, komparatif, atau kontekstual, sehingga pendekatan berbasis keyword saja tidak cukup untuk menangkap keseluruhan sentimen. Namun, keyword-based analysis tetap bernilai sebagai lapisan interpretabilitas dan validasi awal, sekaligus membantu mengidentifikasi komentar dengan sentimen campuran yang relevan untuk proses active learning dan aspect-based sentiment analysis.

```python
# Optional: compile regex biar lebih cepat
NEG_PATTERNS = [re.compile(rf"\b{re.escape(w)}\b") for w in BOOST_WORDS["negative"]]
POS_PATTERNS = [re.compile(rf"\b{re.escape(w)}\b") for w in BOOST_WORDS["positive"]]

def _has_match(patterns, text):
    return any(p.search(text) for p in patterns)

def _normalize_label(label: str) -> str:
    label = label.lower()
    if label in {"positive", "neutral", "negative"}:
        return label
    # fallback mapping (kalau ternyata model keluarnya LABEL_X)
    mapping = {"label_0": "negative", "label_1": "neutral", "label_2": "positive"}
    return mapping.get(label, "neutral")

def analyze_hybrid_sentiment(text: str, override_threshold: float = 0.90):
    # Safety
    if not isinstance(text, str) or not text.strip():
        return "neutral", 0.0

    t = text.lower()

    # A. Prediksi Model (token-level truncation)
    try:
        res = sentiment_pipeline(t, truncation=True, max_length=512)[0]
        label = _normalize_label(res.get("label", "neutral"))
        score = float(res.get("score", 0.0))
    except Exception:
        return "neutral", 0.0

    # B. Keyword Boosting (rule-based guardrail)
    has_neg = _has_match(NEG_PATTERNS, t)
    has_pos = _has_match(POS_PATTERNS, t)

    # Kalau ada sinyal negatif kuat, biasanya lebih ‚Äútegas‚Äù untuk override
    if has_neg:
        if (label != "negative") or (score < override_threshold):
            return "negative", 0.95

    # Positive override hanya jika tidak ada sinyal negatif yang juga kuat
    if has_pos and not has_neg:
        if (label != "positive") or (score < override_threshold):
            return "positive", 0.95

    return label, score
```

### 5. Fine Tuning (Setfit)

```python
# 0) Config path & label set (SINGLE SOURCE OF TRUTH)
CSV_PATH = "Baseline Dataset/baseline_labeling_sample_done.csv"
VALID_LABELS = ["negative", "neutral", "positive"]
SEED = 42

# Optional: stabilin tokenizer
os.environ["TOKENIZERS_PARALLELISM"] = "false"

# 1) Device auto-detect (Azure GPU -> cuda, else cpu)
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device yang dipakai:", device)

# Optional housekeeping
gc.collect()
if device == "cuda":
    try:
        torch.cuda.empty_cache()
    except Exception:
        pass
```

```python
# 2) Load Golden Dataset (manual labels)
try:
    df_gold = pd.read_csv(CSV_PATH)
    print(f"Data training dimuat: {len(df_gold)} sampel")
except FileNotFoundError:
    raise FileNotFoundError(
        f"File '{CSV_PATH}' tidak ditemukan. Pastikan path & nama folder sama persis (case-sensitive di Azure/Linux)."
    )

# 3) Merapikan kolom -> SetFit butuh: text, label
possible_text_cols  = ["text", "clean_text", "comment", "raw_text"]
possible_label_cols = ["label", "gold_label", "sentiment_label", "sentiment"]

text_col  = next((c for c in possible_text_cols  if c in df_gold.columns), None)
label_col = next((c for c in possible_label_cols if c in df_gold.columns), None)

if text_col is None or label_col is None:
    raise ValueError(
        "Kolom teks/label tidak ketemu.\n"
        f"Kolom tersedia: {list(df_gold.columns)}\n"
        f"Harus ada salah satu teks: {possible_text_cols}\n"
        f"dan salah satu label: {possible_label_cols}"
    )

# IMPORTANT FIX (minimal tapi penting):
# Ambil hanya 2 kolom supaya df_gold['text'] pasti Series (hindari error .str pada DataFrame)
df_gold = df_gold[[text_col, label_col]].copy()
df_gold = df_gold.rename(columns={text_col: "text", label_col: "label"})

df_gold["text"]  = df_gold["text"].astype(str).str.strip()
df_gold["label"] = df_gold["label"].astype(str).str.lower().str.strip()
df_gold = df_gold[(df_gold["text"] != "") & (df_gold["label"].isin(VALID_LABELS))].copy()

print("\nDistribusi label (gold):")
print(df_gold["label"].value_counts())
```

```python
# 4) Split train/eval (stratified)
train_df, eval_df = train_test_split(
    df_gold,
    test_size=0.2,
    random_state=SEED,
    stratify=df_gold["label"] if df_gold["label"].nunique() > 1 else None
)

train_dataset = Dataset.from_pandas(train_df.reset_index(drop=True), preserve_index=False)
eval_dataset  = Dataset.from_pandas(eval_df.reset_index(drop=True), preserve_index=False)
print(f"\nTrain: {len(train_dataset)} | Eval: {len(eval_dataset)}")

# 5) Load Base Model
BASE_MODEL = "sentence-transformers/paraphrase-multilingual-mpnet-base-v2"

model = SetFitModel.from_pretrained(
    BASE_MODEL,
    labels=VALID_LABELS
)

# (opsional) pindahin body/head ke device
model.model_body.to(device)
if getattr(model, "model_head", None) is not None:
    try:
        model.model_head.to(device)
    except Exception:
        pass

# 6) Training config (Azure-optimized)
# Update sesuai rekomendasi untuk total ~710 data:
# CPU: batch_size=16, num_epochs=2, num_iterations=6 (balance speed vs quality)
# GPU: tetap boleh lebih tinggi, tapi aku biarin logikamu, cuma aku bikin lebih "waras"

if device == "cuda":
    NUM_ITERATIONS = 12   # lebih realistis (20 sering overkill); bisa naik kalau mau
    BATCH_SIZE     = 32   # kalau OOM turunin ke 16
    NUM_EPOCHS     = 3
else:
    NUM_ITERATIONS = 6
    BATCH_SIZE     = 16
    NUM_EPOCHS     = 2

args = SetFitTrainingArguments(
    batch_size=BATCH_SIZE,
    num_epochs=NUM_EPOCHS,
    num_iterations=NUM_ITERATIONS,
)

# Evaluasi & saving secukupnya
args.eval_strategy = "epoch"
args.save_strategy = "epoch"
args.load_best_model_at_end = True

trainer = SetFitTrainer(
    model=model,
    args=args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)

# 7) Train
print("\nSedang melatih ulang model (Fine-Tuning)...")
print(f"Config -> device={device}, batch_size={BATCH_SIZE}, num_epochs={NUM_EPOCHS}, num_iterations={NUM_ITERATIONS}")

t0 = time.time()
trainer.train()
t1 = time.time()

print(f"\nTraining selesai. Durasi: {(t1 - t0)/60:.2f} menit")
```

```python
# 8) Save model (biar reusable untuk infer & active learning)
SAVE_DIR = "Finetuned_xiaomi_model-Azure"
trainer.model.save_pretrained(SAVE_DIR)
print(f"\nModel tersimpan di: {SAVE_DIR}")
```

```python
# 7) Evaluasi cepat di eval set (classification report + confusion matrix)
print("\nEVALUASI (SET FINE-TUNED):")

eval_texts = eval_df["text"].astype(str).tolist()
y_true = eval_df["label"].astype(str).tolist()

# Predict batch biar cepat
def predict_in_batches(m, texts, batch_size=64):
    preds = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        preds.extend(m(batch))
    return [str(p).lower() for p in preds]

y_pred = predict_in_batches(model, eval_texts, batch_size=64)

print(classification_report(y_true, y_pred, digits=4))

cm = confusion_matrix(y_true, y_pred, labels=["negative", "neutral", "positive"])
disp = ConfusionMatrixDisplay(cm, display_labels=["negative", "neutral", "positive"])
disp.plot(values_format="d")
plt.title("Confusion Matrix - SetFit Fine-Tuned (Eval Set)")
plt.show()
```

```python
# 8) Sanity test contoh manual
print("\nTEST HASIL FINE-TUNING:")
test_cases = [
    "kameranya kek dadu anjit",
    "lebih suka design redmi sebelumnya",
    "tuh bahkan di studio david aja make shizuku wajib sih ini buat debloat yang aneh wkwkwk",
    "seri paling nanggung, mending ambil yang pro",
    "masih pake redmi note 10s aja"
]
preds = model(test_cases)
for t, p in zip(test_cases, preds):
    print(f"Text: '{t}' -> Prediksi: {p}")
```

```python
# --- Compile regex patterns (biar cepat & minim false match) ---
# Word boundary untuk single-word; phrase akan tetap kebaca karena ada spasi di pattern
ASPECT_PATTERNS = {}
for aspect, keywords in ASPECTS_DICT.items():
    patterns = []
    for kw in keywords:
        kw = kw.strip().lower()
        # kalau keyword mengandung spasi atau angka/simbol, boundary tetap aman
        # contoh: "120w", "4k", "value for money"
        patterns.append(re.compile(rf"(?<!\w){re.escape(kw)}(?!\w)"))
    ASPECT_PATTERNS[aspect] = patterns


def tag_aspects(text: str):
    if not isinstance(text, str) or not text.strip():
        return ["umum"]

    t = text.lower()
    tags = []

    for aspect, patterns in ASPECT_PATTERNS.items():
        if any(p.search(t) for p in patterns):
            tags.append(aspect)

    return tags if tags else ["umum"]


# =========================
# 1) Tag aspect
# =========================
# Pastikan df punya clean_text & polarity
if "clean_text" not in df.columns:
    raise ValueError("Kolom `clean_text` tidak ada. Pastikan preprocessing sudah dijalankan.")
if "polarity" not in df.columns:
    raise ValueError("Kolom `polarity` tidak ada. Pastikan sentiment labeling sudah dijalankan.")

df["aspects"] = df["clean_text"].apply(tag_aspects)

# Explode dataset (satu komentar bisa punya banyak aspek)
df_exploded = df.explode("aspects")

# =========================
# 2) Ringkasan sentiment per aspek
# =========================
aspect_summary = (
    df_exploded
    .groupby("aspects")
    .agg(
        total_mentions=("clean_text", "count"),
        avg_sentiment=("polarity", "mean"),
    )
    .sort_values("total_mentions", ascending=False)
)
```

```python
# =========================
# 3) Z-score (salience based on avg_sentiment)
# =========================
mean = aspect_summary["avg_sentiment"].mean()
std = aspect_summary["avg_sentiment"].std()

# Guard: kalau std=0, z-score diset 0 agar aman
if std == 0 or np.isnan(std):
    aspect_summary["z_score"] = 0.0
else:
    aspect_summary["z_score"] = (aspect_summary["avg_sentiment"] - mean) / std

aspect_summary["sentiment_status"] = aspect_summary["z_score"].apply(
    lambda z: "üü¢ Positif" if z > 0.5
    else ("üî¥ Negatif" if z < -0.5 else "üü° Netral")
)

print("SENTIMEN BERDASARKAN FITUR (Z-SCORE):")
display(aspect_summary)
```

```python
# ‚îÄ‚îÄ CONFIG ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
INPUT_TXT = "Output Analisis/Final_Recomendation.txt"
OUTPUT_PDF = "Xiaomi_Note_15_Post_Launch_Report.pdf"

# ‚îÄ‚îÄ LOAD CONTENT ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
if not os.path.exists(INPUT_TXT):
    raise FileNotFoundError(f"File tidak ditemukan: {INPUT_TXT}")

with open(INPUT_TXT, "r", encoding="utf-8") as f:
    raw_text = f.read().strip()

lines = [line.rstrip() for line in raw_text.split("\n")]

# ‚îÄ‚îÄ PDF SETUP ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
doc = SimpleDocTemplate(
    OUTPUT_PDF,
    pagesize=A4,
    leftMargin=2.8*cm,
    rightMargin=2.8*cm,
    topMargin=3.5*cm,     # lebih longgar di atas untuk cover
    bottomMargin=2.5*cm,
)

styles = getSampleStyleSheet()

def add_style(name, **kwargs):
    if name not in styles:
        styles.add(ParagraphStyle(name=name, **kwargs))

# Cover page styles
add_style("CoverMainTitle",   fontSize=26, leading=30, alignment=TA_CENTER, spaceAfter=12, textColor=HexColor('#0D47A1'), fontName='Helvetica-Bold')
add_style("CoverProduct",     fontSize=20, leading=24, alignment=TA_CENTER, spaceAfter=0,  textColor=HexColor('#1565C0'), fontName='Helvetica')

# Content styles
add_style("Section",   fontSize=15.5, leading=19, spaceBefore=16, spaceAfter=7,  textColor=HexColor('#1565C0'), fontName='Helvetica-Bold')
add_style("Subsection",fontSize=13,    leading=16, spaceBefore=12, spaceAfter=6, leftIndent=8,  textColor=HexColor('#424242'), fontName='Helvetica-Bold')
add_style("Body",      fontSize=10.7,  leading=14.5, alignment=TA_JUSTIFY, spaceAfter=7, spaceBefore=3)
add_style("Bullet",    fontSize=10.5,  leading=14, leftIndent=26, bulletIndent=13, spaceAfter=5)

story = []

# ‚îÄ‚îÄ MARKDOWN ‚Üí REPORTLAB CONVERTER ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
def md_to_rl(text):
    text = re.sub(r'\*\*(.+?)\*\*', r'<b>\1</b>', text)
    text = re.sub(r'\*(.+?)\*',      r'<i>\1</i>', text)
    text = text.replace('**', '').replace('*', '')
    return text

# ‚îÄ‚îÄ COVER PAGE (halaman pertama ‚Äì minimalis) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
story.append(Spacer(1, 8*cm))  # jarak dari atas ‚Üí membuat judul terpusat vertikal

story.append(Paragraph(
    "Post-Launch Market Intelligence Report",
    styles["CoverMainTitle"]
))

story.append(Spacer(1, 1.2*cm))

story.append(Paragraph(
    "Xiaomi Note 15 5G",
    styles["CoverProduct"]
))

story.append(Spacer(1, 10*cm))  # jarak bawah agar cover terasa balanced

story.append(PageBreak())   # Pindah ke halaman isi

# ‚îÄ‚îÄ CONTENT PARSER ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
bullet_buffer = []
prev_was_section = False

def flush_bullets():
    if not bullet_buffer:
        return
    items = [ListItem(Paragraph(md_to_rl(txt.strip()), styles["Bullet"])) for txt in bullet_buffer]
    story.append(ListFlowable(
        items,
        bulletType='bullet',
        bulletFontName='Helvetica',
        bulletFontSize=10.8,
        leftPadding=6,
        spaceBetween=3.5
    ))
    bullet_buffer.clear()

for line in lines:
    stripped = line.strip()

    # Skip garis pemisah (---, ‚Ä¢--, ==== , dll) dan baris kosong
    if re.match(r'^[-‚Äî‚îÅ‚Ä¢*_=]{3,}$', stripped) or not stripped:
        flush_bullets()
        story.append(Spacer(1, 5))
        continue

    # Section utama dengan "### "
    if stripped.startswith('### '):
        flush_bullets()
        if prev_was_section:
            story.append(PageBreak())
        else:
            story.append(Spacer(1, 8))

        section_title = stripped[4:].strip()
        story.append(Paragraph(md_to_rl(section_title), styles["Section"]))
        story.append(Spacer(1, 6))
        prev_was_section = True
        continue

    # Subsection (**...** atau ...:)
    if (stripped.startswith('**') and stripped.endswith('**')) or \
       (stripped.endswith(':') and len(stripped) < 50 and not stripped.isupper()):
        flush_bullets()
        clean = stripped.strip('*').strip(':').strip()
        story.append(Paragraph(f"<b>{clean}</b>", styles["Subsection"]))
        story.append(Spacer(1, 5))
        continue

    # Bullet
    if stripped.startswith(('-', '‚Ä¢', '*', '‚Üí', '‚Äì')) or re.match(r'^\s*[-*‚Ä¢‚Üí‚Äì]\s', line):
        bullet_text = re.sub(r'^[-*‚Ä¢‚Üí‚Äì]\s*', '', stripped).strip()
        bullet_buffer.append(bullet_text)
        continue

    # Paragraf biasa
    flush_bullets()
    story.append(Paragraph(md_to_rl(stripped), styles["Body"]))
    story.append(Spacer(1, 4))

flush_bullets()

# ‚îÄ‚îÄ BUILD ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
doc.build(story)

abs_path = os.path.abspath(OUTPUT_PDF)
print(f"PDF selesai ‚Üí {abs_path}")
print(f"Ukuran: {os.path.getsize(OUTPUT_PDF) / 1024:.1f} KB")
```
